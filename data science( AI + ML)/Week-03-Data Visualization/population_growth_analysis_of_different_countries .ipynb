```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
country_data=pd.read_csv("countries.csv")
```

```python
country_data
```

```python
pk=country_data[country_data.country=='Pakistan']
```

```python
pk
```

```python
ind=country_data[country_data.country=='India']
```

```python
ind
```

```python
afg=country_data[country_data.country=='Afghanistan']
```

```python
afg
```

```python
#display population growth by year of  countries
plt.plot(afg.year,afg.population/1000000)#dividing population by 1 million
plt.plot(ind.year, ind.population / 10**6)
plt.plot(pk.year,pk.population/10**6)
plt.legend(['Afghanistan', 'India','Pakistan'])#jo graph phly ha phla label usko mily ga
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
#display population growth by year of  countries by percentage
plt.plot(afg.year,afg.population/afg.population.iloc[0]*100)#dividing population by 1 million
plt.plot(ind.year, ind.population /ind.population.iloc[0]*100)
plt.plot(pk.year,pk.population/pk.population.iloc[0]*100)
plt.legend(['Afghanistan', 'India','Pakistan'])#jo graph phly ha phla label usko mily ga
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
