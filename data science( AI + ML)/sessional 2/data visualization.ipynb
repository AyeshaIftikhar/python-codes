```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
country_data=pd.read_csv("countries.csv")
```

```python
country_data
```

```python
country_data.sort_values(by=['population'],inplace=True)
country_data.tail(50)

```

```python
china=country_data[country_data.country=='China']
```

```python
us=country_data[country_data.country=='United States']
```

```python
india=country_data[country_data.country=='India']
```

```python
indonasia=country_data[country_data.country=='Indonesia']
```

```python
pak=country_data[country_data.country=='Pakistan']
```

```python
brazil=country_data[country_data.country=='Brazil']
```

```python
#display population growth by year of  countries
plt.plot(china.year,china.population/1000000)#dividing population by 1 million
plt.plot(us.year, us.population / 10**6)
plt.plot(india.year,india.population/10**6)
plt.plot(indonasia.year, indonasia.population / 10**6)
plt.plot(pak.year,pak.population/10**6)
plt.plot(brazil.year,brazil.population/10**6)
plt.legend(['China', 'United States','India','Indonasia','Pakistan','Brazil'])#jo graph phly ha phla label usko mily ga
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
